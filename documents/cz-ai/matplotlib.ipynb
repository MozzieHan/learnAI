````python
# ,axes
# figure
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 8), dpi=80)
# x, y
x = range(60)
# y
y_ticks = range(40)
y_shanghai = [random.uniform(15, 18) for i in x]
# 
y_beijing = [random.uniform(1, 3) for i in x]
# 
x_ch = ["11{}".format(i) for i in x]
# 
axes[0].plot(x, y_shanghai, label="")
# plot
axes[1].plot(x, y_beijing, color='r', linestyle='--', label="")
# x,y
# ,
axes[0].set_xticks(x[::5], x_ch[::5])
axes[0].set_yticks(y_ticks[::5])
axes[1].set_xticks(x[::5], x_ch[::5])
axes[1].set_yticks(y_ticks[::5])
# x,y
axes[0].set_xlabel("")
axes[0].set_ylabel("")
axes[1].set_xlabel("")
axes[1].set_ylabel("")
axes[0].set_title("11012")
axes[1].set_title("11012S")
axes[0].legend(loc="best")
axes[1].legend(loc="best")
plt.show()
````

````python
plt.figure(figsize=(20,8), dpi=100)
movie_name = ['3:','','','','','','','','','','']
y = [73853,57767,22354,15969,14839,8725,8716,8318,7916,6764,52222]
x = range(len(movie_name))
plt.bar(x, y, width=0.6, color=['b','r','g','y','c','m','y','k','c','g','g'])
plt.xticks(x, movie_name)
plt.yticks([10000 * i for i in range(9)])
plt.show()
````

````python
plt.figure(figsize=(20,8), dpi=100)
movie_name = ['3:','','']
first_day = [10587.6,10062.5,1275.7]
first_weekend=[36224.9,34479.6,11830]
x = range(len(movie_name))
plt.bar(x, first_day, width=0.2, label="")
plt.bar([i+0.2 for i in x], first_weekend, width=0.2, color="g", label="")
plt.xticks([i+0.1 for i in x], movie_name)
for i in range(len(x)):
    plt.text(x[i], first_day[i]+ 300, first_day[i], ha="center", va="bottom", fontsize=11)
    plt.text(x[i]+0.2, first_weekend[i] + 300, first_weekend[i], ha="center", va="bottom", fontsize=11)
plt.legend(loc='best')
plt.show()
````
